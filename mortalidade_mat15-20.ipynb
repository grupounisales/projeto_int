```python
# Lendo os dataframes

def read_csvs():

  df_mat20 = pd.read_csv('DOMAT20.csv')
  df_mat19 = pd.read_csv('DOMAT19.csv')
  df_mat18 = pd.read_csv('DOMAT18.csv')
  df_mat17 = pd.read_csv('DOMAT17.csv')
  df_mat16 = pd.read_csv('DOMAT16.csv')
  df_mat15 = pd.read_csv('DOMAT15.csv')

  return df_mat20, df_mat19, df_mat18, df_mat17, df_mat16, df_mat15
```

```python
# Unificando os dataframes e gerando o novo arquivo CSV

def union_dataframes(df20, df19, df18, df17, df16, df15):
  
  df_concat = pd.concat([df20, df19, df18, df17, df16, df15])

  return pd.DataFrame(df_concat).to_csv("union.csv", encoding='utf-8', index=False)

```

```python
# Lendo dataset unificado

def select_cols_union():

  df_new = pd.read_csv('union.csv')

  new_csv = {}

  new_csv['DTOBITO'] = df_new['DTOBITO']
  new_csv['NATURAL'] = df_new['NATURAL']
  new_csv['IDADE'] = df_new['IDADE']
  new_csv['RACACOR'] = df_new['RACACOR']
  new_csv['ESTCIV'] = df_new['ESTCIV']
  new_csv['ESC2010'] = df_new['ESC2010']
  new_csv['OCUP'] = df_new['OCUP']
  new_csv['CODMUNRES'] = df_new['CODMUNRES']
  new_csv['OBITOGRAV'] = df_new['OBITOGRAV']
  new_csv['OBITOPUERP'] = df_new['OBITOPUERP']
  new_csv['ASSISTMED'] = df_new['ASSISTMED']

  return pd.DataFrame(new_csv).to_csv("new_cols_union.csv", encoding='utf-8', index=False)

```

```python
# Calculando:
#   % Missing
#   % Vazio
#   % Inconsistência

def calc_perc(df):

  #### Coluna DTOBITO ####

  # Quantidade de campos vazios
  null_dtObito = ((df['DTOBITO'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['DTOBITO']:
    if (len(str(elem)) != 8):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_dtObito = abs(((cont_incons / df.shape[0]) * 100) - null_dtObito)
  

  #### Coluna NATURAL ####

  # Quantidade de campos vazios
  null_natural = ((df['NATURAL'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['NATURAL']:
    if (len(str(elem)) != 3):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_natural = abs(((cont_incons / df.shape[0]) * 100) - null_natural)

  
  #### Coluna IDADE ####

  # Quantidade de campos vazios
  null_idade = ((df['IDADE'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['IDADE']:
    if (len(str(elem)) != 3):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_idade = abs(((cont_incons / df.shape[0]) * 100) - null_idade)

  
  #### Coluna RACACOR ####

  # Quantidade de campos vazios
  null_racacor = ((df['RACACOR'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['RACACOR']:
    if (len(str(elem)) > 1):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_racacor = abs(((cont_incons / df.shape[0]) * 100) - null_racacor)


  #### Coluna ESTCIV ####

  # Quantidade de campos vazios
  null_estciv = ((df['ESTCIV'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['ESTCIV']:
    if (len(str(elem)) > 1):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_estciv = abs(((cont_incons / df.shape[0]) * 100) - null_estciv)


  #### Coluna ESC2010 ####

  # Quantidade de campos vazios
  null_esc2010 = ((df['ESC2010'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['ESC2010']:
    if (len(str(elem)) > 1):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_esc2010 = abs(((cont_incons / df.shape[0]) * 100) - null_esc2010)


  #### Coluna OCUP ####

  # Quantidade de campos vazios
  null_ocup = ((df['OCUP'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['OCUP']:
    if (len(str(elem)) != 6):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_ocup = abs(((cont_incons / df.shape[0]) * 100) - null_ocup)


  #### Coluna CODMUNRES ####

  # Quantidade de campos vazios
  null_codmunres = ((df['CODMUNRES'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['CODMUNRES']:
    if (len(str(elem)) != 7):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_codmunres = abs(((cont_incons / df.shape[0]) * 100) - null_codmunres)


  #### Coluna OBITOGRAV ####

  # Quantidade de campos vazios
  null_obitograv = ((df['OBITOGRAV'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['OBITOGRAV']:
    if (len(str(elem)) > 1):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_obitograv = abs(((cont_incons / df.shape[0]) * 100) - null_obitograv)


  #### Coluna OBITOPUERP ####

  # Quantidade de campos vazios
  null_obitopuerp = ((df['OBITOPUERP'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['OBITOPUERP']:
    if (len(str(elem)) > 1):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_obitopuerp  = abs(((cont_incons / df.shape[0]) * 100) - null_obitopuerp)


  #### Coluna ASSISTMED ####

  # Quantidade de campos vazios
  null_assistmed = ((df['ASSISTMED'].isnull().sum() / df.shape[0]) * 100)

  cont_incons = 0
  for elem in df['ASSISTMED']:
    if (len(str(elem)) > 1):
      cont_incons += 1

  # Quantidade de campos insconsistentes
  incons_assistmed = abs(((cont_incons / df.shape[0]) * 100) - null_assistmed)

  new_csv = {}

  new_csv['%DTOBITO_NULL'] = [f'{null_dtObito:,.2f}%']
  new_csv['%DTOBITO_INCONS'] = [f'{incons_dtObito:,.2f}%']

  new_csv['%NATURAL_NULL'] = [f'{null_natural:,.2f}%']
  new_csv['%NATURAL_INCONS'] = [f'{incons_natural:,.2f}%']
  
  new_csv['%IDADE_NULL'] = [f'{null_idade:,.2f}%']
  new_csv['%IDADE_INCONS'] = [f'{incons_idade:,.2f}%']

  new_csv['%RACACOR_NULL'] = [f'{null_racacor:,.2f}%']
  new_csv['%RACACOR_INCONS'] = [f'{incons_racacor:,.2f}%']

  new_csv['%ESTCIV'] = [f'{null_estciv:,.2f}%']
  new_csv['%ESTCIV'] = [f'{incons_estciv:,.2f}%']

  new_csv['%ESC2010'] = [f'{null_esc2010:,.2f}%']
  new_csv['%ESC2010'] = [f'{incons_esc2010:,.2f}%']

  new_csv['%OCUP'] = [f'{null_ocup:,.2f}%']
  new_csv['%OCUP'] = [f'{incons_ocup:,.2f}%']

  new_csv['%CODMUNRES'] = [f'{null_codmunres:,.2f}%']
  new_csv['%CODMUNRES'] = [f'{incons_codmunres:,.2f}%']

  new_csv['%OBITOGRAV'] = [f'{null_obitograv:,.2f}%']
  new_csv['%OBITOGRAV'] = [f'{incons_obitograv:,.2f}%']

  new_csv['%OBITOPUERP'] = [f'{null_obitopuerp:,.2f}%']
  new_csv['%OBITOPUERP'] = [f'{incons_obitopuerp:,.2f}%']

  new_csv['%ASSISTMED'] = [f'{null_assistmed:,.2f}%']
  new_csv['%ASSISTMED'] = [f'{incons_assistmed:,.2f}%']
  

  # Gerando csv para análise de qualidade  
  df_analise = pd.DataFrame(new_csv).to_excel("analise_qualidade.xlsx", encoding='utf-8', index=False)

```

```python
# Transformando datas

def transform_dtObito(df):

  dt_obitos = []

  for i in range(len(df['DTOBITO'])):
    dt_obitos.append(df['DTOBITO'][i])

  dt_obitos_aux = []

  for data in dt_obitos:
    data_aux = str(data)
    if (len(data_aux) == 7):
      dt_obitos_aux.append('0' + data_aux) 
    else:
      dt_obitos_aux.append(data_aux)

  dt_obito_format = []

  for data in dt_obitos_aux:
    ano = data[4:8] 
    mes = data[2:4]
    dia = data[0:2]
    dt_obito_format.append(ano + '-' + mes + '-' + dia)

  df['DTOBITO'] = pd.to_datetime(dt_obito_format)
```

```python
# Transformando código naturalidade
# Extraindo somente os dígitos correspondentes ao estado

def transf_natural(df):

  arr_natural = []

  for elem in df['NATURAL']:
    if (len(str(elem)) == 5):  
      arr_natural.append(str(elem)[1:3])
    elif (str(elem) == 'nan'):
      arr_natural.append('')
    else:
      arr_natural.append(str(elem)[0:2])

  df['NATURAL'] = arr_natural
```

```python
# Transformando coluna RACACOR

def transf_racacor(df):

  arr_racacor = []

  for raca in df['RACACOR']:
    if (len(str(raca)) == 3 and str(raca) != 'nan'):  
      arr_racacor.append(str(raca)[0])
    elif (str(raca) == 'nan'):
      arr_racacor.append('')

  df['RACACOR'] = arr_racacor
```

```python
# Transformando coluna ESTCIV

def transf_estciv(df):

  arr_estciv = []

  for estciv in df['ESTCIV']:
    if (len(str(estciv)) == 3 and str(estciv) != 'nan'):  
      arr_estciv.append(str(estciv)[0])
    elif (str(estciv) == 'nan'):
      arr_estciv.append('')

  df['ESTCIV'] = arr_estciv
```

```python
# Transformando coluna ESC2010

def transf_esc2010(df):

  arr_esc2010 = []

  for esc2010 in df['ESC2010']:
    if (len(str(esc2010)) == 3 and str(esc2010) != 'nan'):  
      arr_esc2010.append(str(esc2010)[0])
    elif (str(esc2010) == 'nan'):
      arr_esc2010.append('')

  df['ESC2010'] = arr_esc2010
```

```python
# Transformando coluna OBITOGRAV

def transf_obitograv(df):

  arr_obitograv = []

  for obitograv in df['OBITOGRAV']:
    if (len(str(obitograv)) == 3 and str(obitograv) != 'nan'):  
      arr_obitograv.append(str(obitograv)[0])
    elif (str(obitograv) == 'nan'):
      arr_obitograv.append('')

  df['OBITOGRAV'] = arr_obitograv
```

```python
# Transformando coluna OBITOPUERP

def transf_obitopuerp(df):

  arr_obitopuerp = []

  for obitopuerp in df['OBITOPUERP']:
    if (len(str(obitopuerp)) == 3 and str(obitopuerp) != 'nan'):  
      arr_obitopuerp.append(str(obitopuerp)[0])
    elif (str(obitopuerp) == 'nan'):
      arr_obitopuerp.append('')

  df['OBITOPUERP'] = arr_obitopuerp
```

```python
# Transformando coluna ASSISTMED

def transf_assistmed(df):
  
  arr_assistmed = []

  for assistmed in df['ASSISTMED']:
    if (len(str(assistmed)) == 3 and str(assistmed) != 'nan'):  
      arr_assistmed.append(str(assistmed)[0])
    elif (str(assistmed) == 'nan'):
      arr_assistmed.append('')

  df['ASSISTMED'] = arr_assistmed

```

```python
# Gerando novo CSV com os valores das colunas tratados

def create_csv(df):

  new_csv = {}

  new_csv['DTOBITO'] = df['DTOBITO']
  new_csv['NATURAL'] = df['NATURAL']
  new_csv['IDADE'] = df['IDADE']
  new_csv['RACACOR'] = df['RACACOR']
  new_csv['ESTCIV'] = df['ESTCIV']
  new_csv['ESC2010'] = df['ESC2010']
  new_csv['OCUP'] = df['OCUP']
  new_csv['CODMUNRES'] = df['CODMUNRES']
  new_csv['OBITOGRAV'] = df['OBITOGRAV']
  new_csv['OBITOPUERP'] = df['OBITOPUERP']
  new_csv['ASSISTMED'] = df['ASSISTMED']

  return pd.DataFrame(new_csv).to_csv("final_dataset.csv", encoding='utf-8', index=False)
```

```python
import pandas as pd

def main():

  union_dataframes(read_csvs()[0], read_csvs()[1], read_csvs()[2], read_csvs()[3], read_csvs()[4], read_csvs()[5])
  select_cols_union()

  df = pd.read_csv('new_cols_union.csv')

  calc_perc(df)
  transform_dtObito(df)
  transf_natural(df)
  transf_racacor(df)
  transf_estciv(df)
  transf_esc2010(df)
  transf_obitograv(df)
  transf_obitopuerp(df)
  transf_assistmed(df)
  create_csv(df)
  

main()
```
